```python
import string
def clean_text(text):
       # text_nonpunct = [c for c in text if c not in string.punctuation]
        text_nonpunct ="".join([c for c in text if c not in string.punctuation])
        return text_nonpunct
    
r['comclean'] = r['COMMENT'].apply(lambda x: clean_text(x))
r.head()
```
